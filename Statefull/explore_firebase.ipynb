# Firebase Data Explorer
This notebook connects to your Firebase database directly using the service account JSON file. This avoids all `.env` escaping issues with the private key.

```python
import firebase_admin
from firebase_admin import credentials, firestore

# Load directly from the JSON file - avoids all .env escaping issues
SERVICE_ACCOUNT_FILE = r"c:\Users\aryan\Downloads\eduai-e090e-firebase-adminsdk-fbsvc-f547611fa2.json"

try:
    # Reset any previously failed init attempts
    for app in list(firebase_admin._apps.values()):
        firebase_admin.delete_app(app)

    cred = credentials.Certificate(SERVICE_ACCOUNT_FILE)
    firebase_admin.initialize_app(cred)
    print('Firebase initialized successfully!')
    db = firestore.client()
except Exception as e:
    print(f'Error initializing Firebase: {e}')
```

```python
# 1. List all root collections
print("Root Collections in your Firebase Database:\n")
collections = db.collections()
root_collection_names = []
for coll in collections:
    print(f" 📁 {coll.id}")
    root_collection_names.append(coll.id)
```

```python
# 2. Peek into the first 3 documents of each collection to see the data structure
for coll_name in root_collection_names:
    print(f"\n{'='*50}\nCollection: {coll_name}\n{'='*50}")

    docs = db.collection(coll_name).limit(3).stream()

    doc_count = 0
    for doc in docs:
        doc_count += 1
        print(f"\n📄 Document ID: {doc.id}")
        data = doc.to_dict() or {}
        print(f"   Fields (Keys): {list(data.keys())}")
        data_str = str(data)
        if len(data_str) > 200:
            print(f"   Data Preview: {data_str[:200]}...")
        else:
            print(f"   Data: {data_str}")

    if doc_count == 0:
        print("(Collection appears empty or you don't have read access to these docs)")
```
